```python
# Visualization with lineplot
plt.figure(figsize=(12, 5))
sns.lineplot(x='year', y='value', hue='borough', data= df[df['major_category'] == 'Violance Against the Person'], estimator='sum')
plt.title('Visualisasi Kriminalitas Berdasarkan Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Kriminalitas')
plt.legend(title='borough', loc='upper right')
plt.show()
```
