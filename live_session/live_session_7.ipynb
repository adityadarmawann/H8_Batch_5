# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.hist(tips['total_bill'])
plt.title('Total Bill Histogram', size = 15)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.hist(tips['total_bill'], 10, color='green')
plt.title('Total Bill Histogram', size=15)
plt.xlabel('Total Bill', size = 10)
plt.ylabel('Frequency', size = 10)
plt.grid(True)
plt.legend(['Total Bill from Customer'], loc = 0)
plt.savefig('TotalBill_Histogram.png')
plt.show()
```

# Scatter

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Scatter Plot: Total Bill vs Tip', size = 15)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,6))
plt.scatter(tips['size'], tips['tip'])
plt.title('TScatter Plot: Total Bill vs Tip', size=15)
plt.xlabel('Size', size = 10)
plt.ylabel('Tip', size = 10)
plt.grid(True)
plt.savefig('SizeTip_ScatterPlot.png')
plt.show()
```

```python
#using seaborn
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('TScatter Plot: Total Bill vs Tip', size=15)
plt.show()
```

```python
#using seaborn colour
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'day')
plt.title('TScatter Plot: Total Bill vs Tip', size=15)
plt.show()
```

```python
#using seaborn colour and time
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'day', style = 'time')
plt.title('TScatter Plot: Total Bill vs Tip', size=15)
plt.show()
```
